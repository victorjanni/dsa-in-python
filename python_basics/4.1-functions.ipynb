##### Introduction to Functions
Definition:

A function is a block of code that performs a specific task.
Functions help in organizing code, reusing code, and improving readability.



```python
print("hello")
```

```python
# syntax of a function
def function_name(parameters):
    """ Docstring """
    # function body
    return expression
```

```python
age = 10
if age >= 18:
    print("you are an adult")
else: 
    print("you are a minor")    
```

```python
def minor_or_adult(age):
    """ verifying by age input if a person is a minor or an adult"""
    if age >= 18:
        print("you are an adult")
    else: 
        print("you are a minor")    
```

```python
# function calling
minor_or_adult(12)
```

```python
# function with multiple parameters
def multiply(a,b):
    """ multiplying two given numbers"""
    return a * b
result = multiply(3,7)
print(result)
```

```python
# default parameters
def greet(name):
    print(f"hello {name}, welocome to the dsa course")

greet("victor")    
```

```python
def greet(name="techie"):
    print(f"hello {name}, welcome to the ai world")
greet("dhavic")    
```

```python
# variable length arguments
# positional and keywords arguments

def print_numbers(*victor):
    for num in victor:
        print(num)
        
```

```python
print_numbers(1,3,4,56,7,"h","hello")
```

```python
# postional arguments

def print_numbers(*victor):
    for num in victor:
        print(num)
        
```

```python
print_numbers(1,3,4,56,7,"h","hello")
```

```python
# keyword arguments
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_info(name="victor", age=25, city="nellore", profession="Agentic AI engineer")
```

```python
def info(*args, **kwargs):
    for arg in args:
        print(arg)
    for key, value in kwargs.items():
        print(f"{key}: {value}")    
```

```python
info(1,2,3,5,"hi", name="victor", age=25, city="nellore", profession="Agentic AI engineer")
```

```python
# return statement
def add(a, b):
    return a+b
result = add(3,6)
print(result)
```

```python
# Return multiple parameters
def add(a, b):
    return a+b, a*b
addition, multiplication = add(3,6)
```

```python
print(addition)
print(multiplication)
```
