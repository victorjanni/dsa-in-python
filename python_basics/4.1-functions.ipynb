```python
age = 10
if age >= 18:
    print("you are an adult")
else: 
    print("you are a minor")    
```

```python
def minor_or_adult(age):
    """ verifying by age input if a person is a minor or an adult"""
    if age >= 18:
        print("you are an adult")
    else: 
        print("you are a minor")    
```

```python
# function calling
minor_or_adult(12)
```

```python
# function with multiple parameters
def multiply(a,b):
    """ multiplying two given numbers"""
    return a * b
result = multiply(3,7)
print(result)
```
